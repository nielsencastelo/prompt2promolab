```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

src_path = PROJECT_ROOT / "src"
if str(src_path) not in sys.path:
    sys.path.insert(0, str(src_path))

print("PROJECT_ROOT:", PROJECT_ROOT)
```

```python
import json
from pathlib import Path

from prompt2promolab.config import get_outputs_dir
from prompt2promolab.storyboard import load_storyboard
from prompt2promolab.pipelines import load_pipeline
from prompt2promolab.render import render_storyboard
from prompt2promolab.ffmpeg_utils import concat_videos

PROFILE_KEY = "small_gpu"  # options: small_gpu, balanced, quality_14gb_plus
storyboard_path = PROJECT_ROOT / "storyboards" / "nazaapp_30s.json"
storyboard = load_storyboard(storyboard_path)
storyboard
```

### Observação
Em hardware menor, mantenha:
- resolução menor
- menos passos
- poucas cenas por vez

Para produzir variações, altere prompts ou troque somente a primeira e a última cena.


```python
pipe = load_pipeline(PROFILE_KEY)
output_dir = get_outputs_dir() / storyboard["slug"]

clip_paths = render_storyboard(
    pipe=pipe,
    profile_key=PROFILE_KEY,
    storyboard=storyboard,
    output_dir=output_dir,
)

clip_paths
```

```python
final_video = output_dir / f"{storyboard['slug']}_final.mp4"
concat_videos(clip_paths, final_video)
print("Vídeo final salvo em:", final_video)
```
